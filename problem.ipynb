### Part A

(x, y) such that
$$
x = \pm uniform(0, 1)\\
y = \pm uniform(0, 0.5)
$$
and (x, y) lies inside the ellipse. 

```python
from random import *
import matplotlib.pyplot as plt
seed(10e9+7)

count = 0
px = []
py = []
while True:
    x = uniform(0, 1)
    y = uniform(0, 0.5)
    mx = 2*randint(0, 1) - 1
    my = 2*randint(0, 1) - 1
    x = mx*x
    y = my*y
    if x*x + 4*y*y <= 1:
        px.append(x)
        py.append(y)
        count += 1
    if count == 10000000:
        break
plt.hist2d(px, py, bins=50)
plt.colorbar()
plt.show() 
```

### Part B

(x, y) such that
$$
x = uniform(0, \pi)\\
y = uniform(0, e)
$$
and (x, y) lies inside the triangle.

```python
from cmath import pi
from random import *
import matplotlib.pyplot as plt
import math
seed(10e9+7)

count = 0
px = []
py = []
while True:
    x = uniform(0, math.pi)
    y = uniform(0, math.exp(1))
    if y <= 3*math.exp(1)*x/math.pi and y <= -1.5*math.exp(1)*(x-math.pi)/math.pi:
        px.append(x)
        py.append(y)
        count += 1
    if count == 10000000:
        break
plt.hist2d(px, py, bins=50)
plt.colorbar()
plt.show() 
```
